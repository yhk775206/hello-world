```
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
# 1. image 읽기
img = Image.open('Type1_1.jpg')

# 2. HSV 바꾸기
hsv = matplotlib.colors.rgb_to_hsv(img)

h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

# 3. max(v) - min(v)
print ("max(v)-min(v):\t", np.max(v) - np.min(v))

# 4. agerage(v)
print ("avg(v):\t", np.average(v))

plt.imshow(img)
plt.show()
```
